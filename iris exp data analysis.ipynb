```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['species'] = iris.target
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='species', y='sepal length (cm)', data=data)
plt.title('Boxplot of Sepal Length by Species')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='species', y='sepal width (cm)', data=data)
plt.title('Boxplot of Sepal Width by Species')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='species', y='petal length (cm)', data=data)
plt.title('Boxplot of Petal Length by Species')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='species', y='petal width (cm)', data=data)
plt.title('Boxplot of Petal Width by Species')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.pairplot(data, hue='species')
plt.title('Pairplot of Iris Dataset')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='sepal length (cm)', y='sepal width (cm)', hue='species', style='species', s=100)
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='petal length (cm)', y='petal width (cm)', hue='species', style='species', s=100)
plt.title('Petal Length vs Petal Width')
plt.show()
```
